## üßÆ Python Operators for Integers

**What's an Operator?**

An **operator** is a symbol that tells Python to perform an action on values.

Think of it like a command: the `+` operator tells Python "add these numbers together."

---

**Basic Math Operators:**
- `+` Addition: `5 + 3` = `8`
- `-` Subtraction: `10 - 4` = `6`
- `*` Multiplication: `7 * 3` = `21`
- `/` Division: `10 / 2` = `5.0` (gives a float!)
- `//` Integer Division: `10 // 3` = `3` (drops remainder)
- `%` Modulus: `10 % 3` = `1` (gives remainder only)
- `**` Exponent: `2 ** 3` = `8` (2 to the power of 3)

**Order of Operations:** Python follows BIDMAS (Brackets, Indices, Division/Multiplication, Addition/Subtraction)
- `2 + 3 * 4` = `14` (multiplication first)
- `(2 + 3) * 4` = `20` (brackets first)

## üéØLet's use conditionals to compare two userinput numbers

````python
# NOTICE : Input FUNCTION gets user input and returns it as a STRING which is CAST(converted) INTEGER
input1 = int(input())
input2 = int(input())

# STEP 1 : use three if statements
if input1 > input2:
  print("input1 is greater than input2")
if input1 < input2:
  print("input1 is less than input2")
if input1 == input2:
  print("input1 is equal to input2")
# This solution isn't ideal can anyone see why?

# STEP 2 : better control flow, doesn't make any redundant checks
input1 = int(input())
input2 = int(input())

if input1 > input2:
  print("input1 is greater than input2")
elif input1 < input2:
  print("input1 is less than input2")
elif input1 == input2:
  print("input1 is equal to input2")

#STEP 3 :  A better solutions would make no redundant or unecessary checks.
input1 = int(input())
input2 = int(input())

if input1 > input2:
  print("input1 is greater than input2")
elif input1 < input2:
  print("input1 is less than input2")
else:
  print("input1 is equal to input2")

# Assume : I just want to know if they are equal or not.

#STEP 4 :  We can use the OR operator to simplify if
input1 = int(input())
input2 = int(input())

if input1 > input2 or input1 < input2:
  print("input 1 does not equal to input 2")
else:
  print("input1 is equal to input2")

#STEP 5 : Use Not equal to comperator
input1 = int(input())
input2 = int(input())

if input1 != input2:
  print("input 1 does not equal to input 2")
else:
  print("input1 is equal to input2")

#Final
input1 = int(input())
input2 = int(input())

if input1 == input2:
  print("input1 is equal to input2")
else:
  print("input 1 does not equal to input 2")
````

## üîÑ Loops: Repeating Code

**Loops let you run the same code multiple times without copying and pasting.**

Instead of writing `print("Hello")` 100 times, write it once and loop 100 times!

---

## `while` Loop

**Runs code WHILE a condition is True:**

The loop keeps repeating until the condition becomes False.

---

### Basic Structure

```python
while condition:
    # code runs repeatedly
    # until condition is False
```

**Think of it like:** "While I'm hungry, keep eating."

---

### Example: Counting Down

```python
i = 3

while i > 0:
    print(i)
    i = i - 1  # Decrease by 1 each time
```

**What happens:**
1. Check: is `i > 0`? (3 > 0 = True) ‚Üí print 3, decrease to 2
2. Check: is `i > 0`? (2 > 0 = True) ‚Üí print 2, decrease to 1
3. Check: is `i > 0`? (1 > 0 = True) ‚Üí print 1, decrease to 0
4. Check: is `i > 0`? (0 > 0 = False) ‚Üí STOP

**Output:** `3`, `2`, `1`

---

### ‚ö†Ô∏è Warning: Infinite Loops

**If the condition NEVER becomes False, the loop runs FOREVER:**

```python
i = 3
while i > 0:
    print(i)
    # Forgot to decrease i!
    # Loop runs forever printing 3, 3, 3, 3...
```

**Always make sure your condition can eventually become False!**

---

## üí° Key Takeaway

`while` loops repeat code as long as a condition stays True.

Perfect for when you don't know how many times to loop in advance.

## üîÅ `for` Loop: Iterate Through Items

## `range()`: Generating Number Lists

**`range()` creates a sequence of numbers automatically:**

```python
for i in range(3):
```

`range(3)` means "generate numbers from 0 up to (but not including) 3."

```python
range(5)           # 0, 1, 2, 3, 4
range(1, 6)        # 1, 2, 3, 4, 5 (start, stop)

```

## üìã Lists: Complete Guide

---

## Negative Indexing

**Access items from the end of the list using negative numbers:**

```python
fruits = ["apple", "banana", "cherry"]

fruits[-1]   # "cherry" (last item)
fruits[-2]   # "banana" (second to last)
fruits[-3]   # "apple" (third to last)
```

**Think of it as counting backwards from the end.**

---

## Slicing Lists

**Extract a portion of a list:**

```python
fruits = ["apple", "banana", "cherry", "date", "elderberry", "fig"]

fruits[1:4]    # ["banana", "cherry", "date"] (index 1 up to but not including 4)
fruits[:3]     # ["apple", "banana", "cherry"] (from start to index 3)
fruits[3:]     # ["date", "elderberry", "fig"] (from index 3 to end)
fruits[::2]    # ["apple", "cherry", "elderberry"] (every 2nd item)
fruits[::-1]   # ["fig", "elderberry", "date", "cherry", "banana", "apple"] (reversed!)
```

**Syntax:** `list[start:stop:step]`

---

## Adding Items to Lists

### `.append()` - Add single item to end

```python
fruits = ["apple", "banana"]
fruits.append("cherry")
# ["apple", "banana", "cherry"]
```

### `.insert()` - Add at specific position

```python
fruits = ["apple", "cherry"]
fruits.insert(1, "banana")  # Insert at index 1
# ["apple", "banana", "cherry"]
```

---

## Removing Items from Lists

### `.remove()` - Remove first occurrence of value

```python
fruits = ["apple", "banana", "cherry", "banana"]
fruits.remove("banana")  # Removes first "banana"
# ["apple", "cherry", "banana"]
```

### `.pop()` - Remove and return item

```python
fruits = ["apple", "banana", "cherry"]

last = fruits.pop()      # Removes "cherry", returns it
first = fruits.pop(0)    # Removes "apple", returns it
# ["banana"]
```

**`.pop()` with no argument removes the last item.**

---

## Lists Within Lists (Nested Lists)

```python
matrix = [
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9]
]

matrix[0]      # [1, 2, 3] (first row)
matrix[0][1]   # 2 (first row, second item)
```

**Lists can contain other lists!**

---

## `.extend()` - Add Multiple Items

```python
fruits = ["apple", "banana"]
more_fruits = ["cherry", "date"]

fruits.extend(more_fruits)
# ["apple", "banana", "cherry", "date"]
```

**Difference from `.append()`:**
- `.append(more_fruits)` ‚Üí `["apple", "banana", ["cherry", "date"]]` (adds the list itself)
- `.extend(more_fruits)` ‚Üí `["apple", "banana", "cherry", "date"]` (adds each item)

## üìñ Exceptions and Error Handling

### What Are Exceptions?

**Exceptions are Python's way of saying "something went wrong".**

### Common Exception Types

- **ZeroDivisionError** - Dividing by zero
- **TypeError** - Wrong data type for operation
- **FileNotFoundError** - File doesn't exist
- **IndexError** - List index out of range

---

### The try-except Pattern

**Basic structure:**
```python

    # Risky code

    # Handle the error
    print("That's not a number!")
```

---

### Multiple Exception Handlers

```python

    count = int(input("Enter count: "))
    average = total / count
except ValueError:
    print("ERROR: Enter numbers only!")
except ZeroDivisionError:
    print("ERROR: Count cannot be zero!")
```

---

### The else Clause

**Runs ONLY if no exception occurred:**

```python
try:
    score = int(input("Enter score: "))
except ValueError:
    print("Invalid number!")
else:
    print(f"Success! Score: {score}")
```

---

### The finally Clause

**ALWAYS runs - error or not:**

```python
try:
    file = open('data.txt', 'r')
    content = file.read()
except FileNotFoundError:
    print("File not found!")
finally:
    print("Cleanup complete")  # Always executes
```

Use `finally` for cleanup (closing files, logging, etc.)

---

### Complete Structure

```python
try:
    result = risky_operation()
except SpecificError:
    handle_error()
else:
    print("Success!")  # Only if no error
finally:
    print("Done")  # Always runs
```

---

### The raise Keyword

**Manually create exceptions for validation:**


### Re-raising Exceptions

**Catch, log, then propagate:**

## üìñ File I/O: Writing and Reading Text Files

### What is File I/O?

**File I/O (Input/Output) allows programs to save and load data permanently.**

When your program closes, variables disappear. Files let data survive between program runs.

---

### The open() Function

The `open()` function creates a connection to a file on your computer.

**Syntax:** `open(filename, mode)`

**Returns:** A file object you can read from or write to

**Think of it like opening a physical notebook:**
- You open it to a specific page (the file)
- You read or write in it
- You must close it when done

---

### File Modes

**Three main modes:**

- `'w'` - **Write mode** (creates new file or **overwrites** existing content)
- `'r'` - **Read mode** (file must already exist)
- `'a'` - **Append mode** (adds to the end without deleting existing content)

---

### Manual File Handling - The Old Way

**Basic pattern:**
```python
file = open('filename.txt', 'w')
file.write("Some content\n")
file.close()  # MUST close!
```

**The danger:** If you forget `file.close()`, bad things happen.

---

### The Write Mode Trap

**Common beginner mistake:** Using `'w'` mode inside a loop
```python
names = ["Alice", "Bob", "Charlie"]

# WRONG - Each iteration OVERWRITES the entire file!
for name in names:
    file = open('names.txt', 'w')
    file.write(name + '\n')
    file.close()

# Result: Only "Charlie" survives!
```

**Why?** The `'w'` mode deletes everything in the file before writing.

---

### The Fix: Append Mode

**Use `'a'` to add without deleting:**
```python
names = ["Alice", "Bob", "Charlie"]

# CORRECT - Each name gets added to the end
for name in names:
    file = open('names.txt', 'a')
    file.write(name + '\n')
    file.close()

# Result: All three names are saved!
```

---

### More Efficient: Open Once

**Instead of opening/closing in every loop iteration:**
```python
names = ["Alice", "Bob", "Charlie"]

file = open('names.txt', 'w')  # Open once
for name in names:
    file.write(name + '\n')
file.close()  # Close once
```

**Better performance, but still has a critical flaw...**

---

### The Problem: Memory Leaks

**What is a memory leak?**

A memory leak happens when your program uses resources (like an open file) but never releases them back to the system.

**With files:**
- Each `open()` reserves system memory
- If you never `close()`, that memory stays reserved
- Multiple unclosed files = wasted memory
- In long-running programs, this adds up!

**Worse:** If an error occurs between `open()` and `close()`, the file NEVER closes.
```python
file = open('data.txt', 'w')
file.write("Starting...\n")
result = 10 / 0  # Error! Program crashes
file.close()  # This line NEVER runs
# File stays open, memory leaked!
```

---

### The Solution: with Statement

**The `with` statement automatically closes files - even if errors occur.**
```python
with open('names.txt', 'w') as file:
    file.write("Content here\n")
# File automatically closed here!
```

**Benefits:**
- No need to remember `file.close()`
- Guaranteed cleanup, even during errors
- Cleaner, more readable code
- Industry standard practice

---

### Why with Prevents Memory Leaks

**The `with` statement is a context manager** - it ensures cleanup always happens.
```python
with open('data.txt', 'w') as file:
    file.write("Some data\n")
    result = 10 / 0  # Error occurs!
    file.write("Never runs")
# File STILL closes automatically despite error!
```

**No memory leak** - the file connection is properly released back to the system.

---

### Complete Example: Saving Names
```python
# Collect names from user
names = []
for i in range(3):
    name = input("Enter a name: ")
    names.append(name)

# Save to file - THE RIGHT WAY
with open('names.txt', 'w') as file:
    for name in names:
        file.write(name + '\n')

# Read them back
with open('names.txt', 'r') as file:
    content = file.read()
    print(content)
```

---

### Reading Files

**Three ways to read:**
```python
# 1. Read entire file as one string
with open('names.txt', 'r') as file:
    content = file.read()

# 2. Read line by line (memory efficient)
with open('names.txt', 'r') as file:
    for line in file:
        print(line.strip())  # .strip() removes \n

# 3. Read all lines into a list
with open('names.txt', 'r') as file:
    lines = file.readlines()  # Returns ['Alice\n', 'Bob\n', ...]
```

---

### Handling Missing Files

**Reading a non-existent file causes `FileNotFoundError`:**
```python
try:
    with open('missing.txt', 'r') as file:
        content = file.read()
except FileNotFoundError:
    print("File doesn't exist!")
    content = ""
```

**Combine with error handling for robust code.**

---

### Best Practices Summary

‚úÖ **ALWAYS use `with` statement** - prevents memory leaks
‚úÖ **Use `'a'` for adding to files** - not `'w'` in loops  
‚úÖ **Use try-except** - handle `FileNotFoundError`  
‚úÖ **Use `.strip()`** - remove newline characters when reading  
‚úÖ **Remember:** `'w'` deletes everything, `'a'` adds to end

---

### Quick Reference
```python

    file.write("Content\n")

# Appending (adds to end)
with open('file.txt', 'a') as file:
    file.write("More content\n")

# Reading
with open('file.txt', 'r') as file:
    content = file.read()
```

**Think of `with` as an automatic door** - it closes behind you no matter what happens inside.
